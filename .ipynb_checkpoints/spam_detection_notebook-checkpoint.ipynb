## 📂 Étape 2 — Chargement des données

### Qu'est-ce que le dataset SMS Spam Collection ?
- **5 572 SMS** réels collectés au Royaume-Uni et Singapour
- Chaque SMS est étiqueté **ham** (légitime) ou **spam**
- Format : fichier TSV (séparé par des tabulations `\t`)

### Pourquoi `sep='\t'` et pas `sep=','` ?
Le fichier n'est **pas un CSV** (virgule) mais un **TSV** (tabulation).  
Si on utilise le mauvais séparateur, tout le contenu se retrouve dans une seule colonne.


````
# ── Chargement du dataset ──────────────────────────────────
df = pd.read_csv(
    "data/SMSSpamCollection",
    sep="\t",                          # séparateur = tabulation
    header=None,                       # pas de ligne d'en-tête dans le fichier
    names=["label", "message"],        # on nomme les colonnes manuellement
    encoding="latin-1"                 # encodage du fichier
)

print(f"✅ Dataset chargé : {df.shape[0]} lignes × {df.shape[1]} colonnes")
print(f"   Colonnes : {list(df.columns)}")
print()
print("── Aperçu des 5 premières lignes ──────────────")
df.head()
````

````
# ── Informations générales ─────────────────────────────────
print("── Informations sur le dataset ────────────────")
print(df.info())
print()
print("── Valeurs manquantes ? ───────────────────────")
print(df.isnull().sum())
print()
print("── Répartition des classes ────────────────────")
print(df["label"].value_counts())
print()
print(f"   Ratio spam : {df['label'].value_counts()['spam'] / len(df) * 100:.1f}%")
print(f"   Ratio ham  : {df['label'].value_counts()['ham']  / len(df) * 100:.1f}%")
````

## 📊 Étape 3 — Exploration des données (EDA)

### Qu'est-ce que l'EDA (Exploratory Data Analysis) ?
L'EDA consiste à **explorer et comprendre** les données avant de construire le modèle.  
On cherche à répondre aux questions :
- Y a-t-il un déséquilibre entre les classes ?
- Les spams sont-ils plus longs que les messages normaux ?
- Quels mots reviennent le plus souvent ?

### 📌 Statistique clé à retenir
> Si une classe est beaucoup plus rare que l'autre (ex: 13% spam vs 87% ham),  
> le modèle doit être évalué avec **F1-score** et **ROC-AUC**, pas seulement l'accuracy.


````
# ── Ajout de colonnes statistiques ────────────────────────
df["label_num"]  = df["label"].map({"ham": 0, "spam": 1})
df["msg_length"] = df["message"].apply(len)
df["word_count"] = df["message"].apply(lambda x: len(x.split()))
df["has_number"] = df["message"].apply(lambda x: bool(re.search(r'\d', x))).astype(int)
df["has_url"]    = df["message"].apply(lambda x: bool(re.search(r'http|www|\.com', x, re.I))).astype(int)
df["upper_ratio"]= df["message"].apply(lambda x: sum(1 for c in x if c.isupper()) / (len(x)+1))

print("── Statistiques par classe ────────────────────")
stats = df.groupby("label")[["msg_length", "word_count", "has_number", "has_url", "upper_ratio"]].mean().round(3)
print(stats)
````

````
# ── Visualisation 1 : Déséquilibre des classes ─────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle("Étape 3 — Exploration des données SMS", fontsize=14, fontweight="bold")

# Graphique 1 : Répartition ham vs spam
counts = df["label"].value_counts()
colors = ["#378ADD", "#D85A30"]
bars = axes[0].bar(counts.index, counts.values, color=colors, alpha=0.85, width=0.5, edgecolor="white")
axes[0].set_title("Répartition Ham vs Spam")
axes[0].set_ylabel("Nombre de messages")
for bar, val in zip(bars, counts.values):
    axes[0].text(bar.get_x() + bar.get_width()/2, val + 30,
                 f"{val}\n({val/len(df)*100:.1f}%)", ha="center", fontsize=11, fontweight="bold")
axes[0].set_ylim(0, max(counts.values) * 1.15)

# Graphique 2 : Distribution longueur des messages
df[df["label"]=="ham"]["msg_length"].plot(
    kind="hist", bins=50, ax=axes[1], color="#378ADD", alpha=0.7, label="Ham")
df[df["label"]=="spam"]["msg_length"].plot(
    kind="hist", bins=50, ax=axes[1], color="#D85A30", alpha=0.7, label="Spam")
axes[1].set_title("Longueur des messages (nb caractères)")
axes[1].set_xlabel("Nombre de caractères")
axes[1].legend()

# Graphique 3 : Boxplot nombre de mots
df.boxplot(column="word_count", by="label", ax=axes[2],
           boxprops=dict(color="#378ADD"),
           medianprops=dict(color="#D85A30", linewidth=2))
axes[2].set_title("Nombre de mots par message")
axes[2].set_xlabel("Classe")
axes[2].set_ylabel("Nombre de mots")
plt.suptitle("")

plt.tight_layout()
plt.savefig("output/01_eda.png", dpi=150, bbox_inches="tight")
plt.show()
print("\n📌 Interprétation :")
print("   → Les spams sont EN MOYENNE plus longs que les ham")
print(f"   → Spam : {df[df['label']=='spam']['msg_length'].mean():.0f} caractères en moyenne")
print(f"   → Ham  : {df[df['label']=='ham']['msg_length'].mean():.0f} caractères en moyenne")
````

````
# ── Visualisation 2 : Mots les plus fréquents ──────────────
from collections import Counter

def top_words(df_subset, n=15):
    all_words = " ".join(df_subset["message"].str.lower()).split()
    # Supprimer la ponctuation
    all_words = [w.strip(string.punctuation) for w in all_words if len(w) > 2]
    return Counter(all_words).most_common(n)

top_spam = top_words(df[df["label"]=="spam"])
top_ham  = top_words(df[df["label"]=="ham"])

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle("Top 15 mots les plus fréquents", fontsize=13, fontweight="bold")

# Spam words
words_s, counts_s = zip(*top_spam)
axes[0].barh(words_s[::-1], counts_s[::-1], color="#D85A30", alpha=0.85)
axes[0].set_title("Mots fréquents dans les SPAM")
axes[0].set_xlabel("Fréquence")

# Ham words
words_h, counts_h = zip(*top_ham)
axes[1].barh(words_h[::-1], counts_h[::-1], color="#378ADD", alpha=0.85)
axes[1].set_title("Mots fréquents dans les HAM")
axes[1].set_xlabel("Fréquence")

plt.tight_layout()
plt.savefig("output/02_top_words.png", dpi=150, bbox_inches="tight")
plt.show()
print("\n📌 Interprétation :")
print("   → Les spams utilisent : 'free', 'call', 'txt', 'win', 'prize'")
print("   → Les ham utilisent   : 'I', 'you', 'to', 'the', 'a' (langage naturel)")
````

## 🧹 Étape 4 — Prétraitement du texte (NLTK)

### Pourquoi prétraiter le texte ?
Un modèle ML ne comprend pas le texte brut — il a besoin de **nombres**.  
Avant de vectoriser, on nettoie le texte pour :
1. **Réduire le bruit** (ponctuation, chiffres, majuscules)
2. **Regrouper les variantes** d'un même mot (`running` → `run`)
3. **Supprimer les mots inutiles** (`the`, `is`, `a`...)

### Les 5 étapes du prétraitement
```
Texte brut → Minuscules → Sans chiffres → Sans ponctuation → Sans stopwords → Stemming
```

### 📌 Stemming vs Lemmatisation
| | Stemming | Lemmatisation |
|--|--|--|
| Méthode | Coupe les suffixes | Utilise un dictionnaire |
| Exemple | `running` → `run` | `running` → `run` |
| Vitesse | ⚡ Rapide | 🐢 Plus lent |
| Précision | Approximatif | Précis |
| Utilisation | NLP classique | Chatbots, traduction |


````
# ── Fonction de prétraitement ──────────────────────────────
stemmer    = PorterStemmer()
stop_words = set(stopwords.words("english"))

def preprocess(text: str) -> str:
    """
    Nettoie un message SMS en 5 étapes :
    1. Convertir en minuscules
    2. Supprimer les chiffres
    3. Supprimer la ponctuation
    4. Supprimer les stopwords
    5. Appliquer le stemming
    """
    # Étape 1 : minuscules
    text = text.lower()
    # Étape 2 : supprimer les chiffres
    text = re.sub(r"\d+", "", text)
    # Étape 3 : supprimer la ponctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    # Étape 4 + 5 : tokeniser, filtrer stopwords, stemmer
    tokens = [stemmer.stem(t) for t in text.split()
              if t not in stop_words and len(t) > 2]
    return " ".join(tokens)

# ── Démonstration étape par étape ──────────────────────────
exemple = "FREE entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005!"

print("📩 Message original :")
print(f"   '{exemple}'")
print()
print("🔄 Étapes du prétraitement :")
print(f"   1. Minuscules   : '{exemple.lower()}'")
e2 = re.sub(r"\d+", "", exemple.lower())
print(f"   2. Sans chiffres: '{e2}'")
e3 = e2.translate(str.maketrans("", "", string.punctuation))
print(f"   3. Sans ponct.  : '{e3}'")
e4 = [t for t in e3.split() if t not in stop_words and len(t) > 2]
print(f"   4. Sans stopw.  : {e4}")
e5 = [stemmer.stem(t) for t in e4]
print(f"   5. Après stem.  : {e5}")
print()
print(f"✅ Résultat final : '{preprocess(exemple)}'")
````

````
# ── Application sur tout le dataset ───────────────────────
df["clean_message"] = df["message"].apply(preprocess)

print("── Avant / Après prétraitement ────────────────")
for i in [2, 9, 12]:   # un spam, un ham, un spam
    print(f"[{df['label'].iloc[i].upper()}]")
    print(f"  Avant : {df['message'].iloc[i][:80]}")
    print(f"  Après : {df['clean_message'].iloc[i]}")
    print()

print(f"✅ Colonne 'clean_message' ajoutée — {len(df)} messages traités")

# Longueur moyenne après nettoyage
df["clean_length"] = df["clean_message"].apply(len)
print(f"\n── Réduction de la taille du texte ───────────")
print(f"   Longueur moyenne AVANT : {df['msg_length'].mean():.0f} caractères")
print(f"   Longueur moyenne APRÈS : {df['clean_length'].mean():.0f} caractères")
print(f"   Réduction              : {(1 - df['clean_length'].mean()/df['msg_length'].mean())*100:.0f}%")
````

## 🔢 Étape 5 — TF-IDF : Transformer le texte en chiffres

### Pourquoi TF-IDF ?
Un modèle ML ne peut pas lire du texte — il faut convertir chaque message  
en un **vecteur de nombres**.

### 📐 La formule TF-IDF
$$TF\text{-}IDF(t, d) = TF(t,d) \times IDF(t)$$

| Composante | Formule | Signification |
|---|---|---|
| **TF** (Term Frequency) | nb_occurrences / nb_mots_doc | Fréquence du mot dans CE message |
| **IDF** (Inverse Document Frequency) | log(nb_docs / nb_docs_avec_mot) | Rareté du mot dans TOUS les messages |
| **TF-IDF** | TF × IDF | Importance du mot dans CE message par rapport à tous les autres |

### 📌 Intuition
- Le mot **"free"** apparaît dans beaucoup de spams → TF élevé dans spams, IDF faible globalement
- Le mot **"the"** apparaît partout → IDF très faible → score TF-IDF proche de 0
- Un mot rare et présent dans un message → score TF-IDF élevé → très discriminant


````
# ── Démonstration TF-IDF sur 5 messages ───────────────────
demo_messages = df["clean_message"].head(5).tolist()

vectorizer_demo = TfidfVectorizer(max_features=20)
tfidf_demo = vectorizer_demo.fit_transform(demo_messages)

df_tfidf_demo = pd.DataFrame(
    tfidf_demo.toarray(),
    columns=vectorizer_demo.get_feature_names_out()
)
df_tfidf_demo.index = [f"Message {i+1}" for i in range(5)]

print("── Matrice TF-IDF (5 messages × 20 termes) ───")
print("   Chaque valeur = importance du mot dans ce message")
print()
df_tfidf_demo.round(3)
````

````
# ── Visualisation de la matrice TF-IDF ────────────────────
plt.figure(figsize=(14, 3))
sns.heatmap(df_tfidf_demo, annot=True, fmt=".2f", cmap="YlOrRd",
            linewidths=0.5, cbar_kws={"label": "Score TF-IDF"})
plt.title("Matrice TF-IDF — 5 premiers messages (valeur 0 = mot absent)", fontsize=12)
plt.xlabel("Termes")
plt.ylabel("Messages")
plt.tight_layout()
plt.savefig("output/03_tfidf_matrix.png", dpi=150, bbox_inches="tight")
plt.show()
print("📌 Plus la couleur est foncée, plus le mot est IMPORTANT dans ce message")
````

## 🧠 Étape 6 — Modèle Naive Bayes

### Pourquoi Naive Bayes pour le spam ?
C'est l'algorithme **classique** pour la classification de texte :
- Rapide à entraîner
- Fonctionne bien avec peu de données
- Basé sur les probabilités → très interprétable

### 📐 La formule de Bayes
$$P(Spam | message) = \frac{P(message | Spam) \times P(Spam)}{P(message)}$$

**En clair :** étant donné les mots du message, quelle est la probabilité que ce soit un spam ?

### Pourquoi "Naive" ?
Il suppose que chaque mot est **indépendant** des autres — ce qui est faux en réalité  
("free prize" ≠ "free" + "prize" indépendamment), mais ça marche quand même très bien !

### 🔗 Pipeline = TF-IDF + Naive Bayes enchaînés
```
Texte propre → TF-IDF (vecteur de nombres) → Naive Bayes (prédiction)
```


````
# ── Préparation des données ────────────────────────────────
X = df["clean_message"]
y = df["label_num"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,      # 80% entraînement, 20% test
    random_state=42,     # reproductibilité
    stratify=y           # même proportion spam/ham dans train et test
)

print("── Division du dataset ────────────────────────")
print(f"   Total    : {len(df)} messages")
print(f"   Train    : {len(X_train)} messages ({len(X_train)/len(df)*100:.0f}%)")
print(f"   Test     : {len(X_test)} messages  ({len(X_test)/len(df)*100:.0f}%)")
print()
print("── Répartition spam/ham dans le test set ──────")
test_counts = y_test.value_counts()
print(f"   Ham  : {test_counts[0]}")
print(f"   Spam : {test_counts[1]}")
````

````
# ── Construction et entraînement du pipeline ──────────────
pipeline_nb = Pipeline([
    ("tfidf", TfidfVectorizer(
        max_features=5000,   # garder les 5000 termes les plus importants
        ngram_range=(1, 2),  # unigrammes ET bigrammes ("free prize")
        min_df=2,            # ignorer les termes qui apparaissent < 2 fois
        sublinear_tf=True    # utiliser log(TF) pour réduire l'impact des mots très fréquents
    )),
    ("clf", MultinomialNB(
        alpha=0.1            # lissage de Laplace (évite les probabilités nulles)
    ))
])

# Entraînement
pipeline_nb.fit(X_train, y_train)

print("✅ Pipeline entraîné !")
print()
print("── Architecture du pipeline ───────────────────")
print("   Étape 1 : TfidfVectorizer")
print(f"             → {pipeline_nb.named_steps['tfidf'].max_features} features max")
print(f"             → n-grams : {pipeline_nb.named_steps['tfidf'].ngram_range}")
print("   Étape 2 : MultinomialNB")
print(f"             → alpha (lissage) : {pipeline_nb.named_steps['clf'].alpha}")
````

## 📈 Étape 7 — Évaluation des performances

### Métriques importantes pour un détecteur de spam

| Métrique | Formule | Ce qu'elle mesure |
|---|---|---|
| **Accuracy** | (VP + VN) / Total | % de bonnes prédictions globales |
| **Précision** | VP / (VP + FP) | Parmi les spams détectés, combien sont vrais ? |
| **Recall** | VP / (VP + FN) | Parmi les vrais spams, combien sont détectés ? |
| **F1-Score** | 2 × (P × R) / (P + R) | Équilibre précision/recall |
| **ROC-AUC** | Aire sous la courbe ROC | Performance globale, robuste au déséquilibre |

### 📌 Quelle métrique privilégier ?
- Pour le spam : le **Recall** est crucial → mieux vaut classer un ham en spam que rater un spam
- L'**Accuracy** seule est trompeuse avec des classes déséquilibrées (87% ham → dire "tout est ham" donne 87% accuracy !)

### La matrice de confusion
```
                 Prédit Ham    Prédit Spam
Réel Ham    │  Vrai Négatif  │ Faux Positif │  ← Ham classé en Spam (embêtant)
Réel Spam   │  Faux Négatif  │ Vrai Positif │  ← Spam classé en Ham (dangereux !)
```


````
# ── Prédictions sur le test set ───────────────────────────
y_pred_nb = pipeline_nb.predict(X_test)
y_prob_nb = pipeline_nb.predict_proba(X_test)[:, 1]

acc = accuracy_score(y_test, y_pred_nb)
auc = roc_auc_score(y_test, y_prob_nb)

print("=" * 50)
print("  RÉSULTATS — Naive Bayes + TF-IDF")
print("=" * 50)
print(f"  Accuracy  : {acc:.4f}  ({acc*100:.2f}%)")
print(f"  ROC-AUC   : {auc:.4f}  ({auc*100:.2f}%)")
print()
print("── Rapport de classification ──────────────────")
print(classification_report(y_test, y_pred_nb, target_names=["Ham", "Spam"]))
````

````
# ── Matrice de confusion ───────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle("Étape 7 — Évaluation du modèle Naive Bayes", fontsize=13, fontweight="bold")

# Matrice de confusion
cm = confusion_matrix(y_test, y_pred_nb)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Ham (prédit)", "Spam (prédit)"],
            yticklabels=["Ham (réel)", "Spam (réel)"],
            linewidths=0.5, ax=axes[0], cbar=False,
            annot_kws={"size": 14, "weight": "bold"})
axes[0].set_title("Matrice de confusion")
tn, fp, fn, tp = cm.ravel()
axes[0].text(0.5, -0.15,
    f"VN={tn}  FP={fp}  FN={fn}  VP={tp}",
    transform=axes[0].transAxes, ha="center", fontsize=10, color="gray")

# Courbe ROC
fpr, tpr, _ = roc_curve(y_test, y_prob_nb)
axes[1].plot(fpr, tpr, color="#378ADD", lw=2.5,
             label=f"Naive Bayes (AUC = {auc:.3f})")
axes[1].plot([0, 1], [0, 1], "k--", alpha=0.4, label="Aléatoire (AUC = 0.5)")
axes[1].fill_between(fpr, tpr, alpha=0.08, color="#378ADD")
axes[1].set_xlabel("Taux de faux positifs (FPR)")
axes[1].set_ylabel("Taux de vrais positifs (TPR)")
axes[1].set_title("Courbe ROC")
axes[1].legend(loc="lower right")
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig("output/04_evaluation.png", dpi=150, bbox_inches="tight")
plt.show()

print(f"\n📌 Interprétation de la matrice :")
print(f"   ✅ Vrais Négatifs (VN)  : {tn} ham correctement classés")
print(f"   ✅ Vrais Positifs (VP)  : {tp} spams correctement détectés")
print(f"   ⚠️  Faux Positifs (FP)  : {fp} ham classés comme spam")
print(f"   🚨 Faux Négatifs (FN)  : {fn} spams non détectés ← le plus dangereux !")
````

## 🏆 Étape 8 — Comparaison de modèles

### Pourquoi comparer plusieurs modèles ?
Il n'existe pas de modèle "universellement meilleur".  
On compare pour choisir le meilleur sur NOS données.

| Modèle | Type | Avantage |
|---|---|---|
| **Naive Bayes** | Probabiliste | Rapide, interprétable, bon sur texte |
| **Logistic Regression** | Linéaire | Stable, probabilités calibrées |
| **Linear SVC** | Marge maximale | Souvent le plus précis sur texte |


````
# ── Définition des 3 modèles ───────────────────────────────
models = {
    "Naive Bayes": pipeline_nb,   # déjà entraîné
    "Logistic Regression": Pipeline([
        ("tfidf", TfidfVectorizer(max_features=5000, ngram_range=(1,2), sublinear_tf=True)),
        ("clf",   LogisticRegression(max_iter=1000, C=1.0, random_state=42))
    ]),
    "Linear SVC": Pipeline([
        ("tfidf", TfidfVectorizer(max_features=5000, ngram_range=(1,2), sublinear_tf=True)),
        ("clf",   LinearSVC(C=1.0, random_state=42))
    ]),
}

# ── Entraînement & évaluation ──────────────────────────────
results = {}
print("⏳ Entraînement en cours...")

for name, model in models.items():
    if name != "Naive Bayes":
        model.fit(X_train, y_train)

    pred     = model.predict(X_test)
    cv_f1    = cross_val_score(model, X, y, cv=5, scoring="f1", n_jobs=-1)

    results[name] = {
        "Accuracy":    round(accuracy_score(y_test, pred), 4),
        "F1-Score":    round(cv_f1.mean(), 4),
        "F1-Std":      round(cv_f1.std(), 4),
        "Precision":   round(classification_report(y_test, pred, output_dict=True)["1"]["precision"], 4),
        "Recall":      round(classification_report(y_test, pred, output_dict=True)["1"]["recall"], 4),
    }
    print(f"   ✅ {name}")

df_results = pd.DataFrame(results).T
print()
print("── Tableau comparatif ─────────────────────────")
df_results
````

````
# ── Visualisation comparaison ──────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle("Étape 8 — Comparaison des modèles", fontsize=13, fontweight="bold")

metrics   = ["Accuracy", "F1-Score", "Precision", "Recall"]
colors    = ["#378ADD", "#1D9E75", "#D85A30", "#BA7517"]
x         = np.arange(len(df_results.index))
width     = 0.2

for i, (metric, color) in enumerate(zip(metrics, colors)):
    axes[0].bar(x + i*width, df_results[metric], width, label=metric, color=color, alpha=0.85)

axes[0].set_xticks(x + width * 1.5)
axes[0].set_xticklabels(df_results.index, rotation=10)
axes[0].set_ylim(0.88, 1.02)
axes[0].set_ylabel("Score")
axes[0].set_title("Métriques par modèle")
axes[0].legend(loc="lower right", fontsize=9)
axes[0].grid(axis="y", alpha=0.3)

# F1 avec écart-type (barres d'erreur)
axes[1].bar(df_results.index, df_results["F1-Score"],
            yerr=df_results["F1-Std"], capsize=8,
            color=["#378ADD", "#1D9E75", "#D85A30"], alpha=0.85, width=0.4)
axes[1].set_ylim(0.88, 1.02)
axes[1].set_ylabel("F1-Score (CV 5-fold)")
axes[1].set_title("F1-Score avec intervalle de confiance")
axes[1].grid(axis="y", alpha=0.3)
for i, (val, std) in enumerate(zip(df_results["F1-Score"], df_results["F1-Std"])):
    axes[1].text(i, val + std + 0.003, f"{val:.4f}", ha="center", fontsize=10, fontweight="bold")

plt.tight_layout()
plt.savefig("output/05_model_comparison.png", dpi=150, bbox_inches="tight")
plt.show()

best = df_results["F1-Score"].idxmax()
print(f"\n🏆 Meilleur modèle : {best} (F1 = {df_results.loc[best, 'F1-Score']:.4f})")
````

## ⚙️ Étape 9 — Optimisation avec GridSearchCV

### Qu'est-ce qu'un hyperparamètre ?
Un hyperparamètre est un **réglage du modèle** qu'on fixe AVANT l'entraînement.  
Ex : `alpha=0.1` dans Naive Bayes, `max_features=5000` dans TF-IDF.

### GridSearchCV — Comment ça marche ?
Il teste **toutes les combinaisons** possibles avec une **validation croisée (k-fold)** :

```
Pour chaque combinaison de paramètres :
  → Diviser les données en 5 parties (folds)
  → Entraîner sur 4 parties, tester sur 1
  → Répéter 5 fois → calculer le score moyen
→ Garder la combinaison avec le meilleur score
```

### 📌 Validation croisée (Cross-Validation)
Évite le **surapprentissage** (overfitting) — le modèle ne "mémorise" pas le jeu de test.


````
# ── GridSearchCV ───────────────────────────────────────────
param_grid = {
    "tfidf__max_features": [3000, 5000],        # nombre de termes TF-IDF
    "tfidf__ngram_range":  [(1, 1), (1, 2)],    # unigrammes vs bigrammes
    "clf__alpha":          [0.01, 0.1, 0.5, 1.0] # lissage Naive Bayes
}

# Nombre total de combinaisons
n_combinations = 2 * 2 * 4
n_folds = 5
print(f"⏳ GridSearch : {n_combinations} combinaisons × {n_folds} folds = {n_combinations*n_folds} entraînements")
print()

gs = GridSearchCV(
    pipeline_nb,
    param_grid,
    cv=5,
    scoring="f1",
    n_jobs=-1,       # utiliser tous les cœurs CPU
    verbose=1
)
gs.fit(X_train, y_train)

print(f"\n✅ Optimisation terminée !")
print(f"   Meilleurs paramètres : {gs.best_params_}")
print(f"   Meilleur F1 (CV)     : {gs.best_score_:.4f}")
````

````
# ── Résultats détaillés du GridSearch ─────────────────────
best_model  = gs.best_estimator_
y_pred_best = best_model.predict(X_test)

print("── Modèle optimisé vs modèle de base ──────────")
print(f"   Accuracy AVANT : {accuracy_score(y_test, y_pred_nb):.4f}")
print(f"   Accuracy APRÈS : {accuracy_score(y_test, y_pred_best):.4f}")
print()

# Tableau des 8 meilleurs résultats
cv_results = pd.DataFrame(gs.cv_results_)
top8 = cv_results.nlargest(8, "mean_test_score")[
    ["param_tfidf__max_features", "param_tfidf__ngram_range",
     "param_clf__alpha", "mean_test_score", "std_test_score"]
].round(4)
top8.columns = ["max_features", "ngram_range", "alpha", "F1 moyen", "F1 std"]
top8.index = range(1, len(top8)+1)
print("── Top 8 combinaisons ─────────────────────────")
top8
````

## 🚀 Étape 10 — Prédiction en temps réel & Sauvegarde

### On sauvegarde le modèle avec `joblib`
`joblib.dump()` sérialise le pipeline complet (TF-IDF + Naive Bayes) dans un fichier `.pkl`.  
Il pourra être rechargé dans Flask sans ré-entraînement.


````
# ── Top features discriminants ─────────────────────────────
vectorizer_final = best_model.named_steps["tfidf"]
clf_final        = best_model.named_steps["clf"]
feature_names    = vectorizer_final.get_feature_names_out()
log_probs        = clf_final.feature_log_prob_[1] - clf_final.feature_log_prob_[0]

top_n = 20
top_spam_idx = np.argsort(log_probs)[-top_n:][::-1]
top_ham_idx  = np.argsort(log_probs)[:top_n]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle("Top 20 mots les plus discriminants (modèle optimisé)", fontsize=13, fontweight="bold")

axes[0].barh(feature_names[top_spam_idx][::-1], log_probs[top_spam_idx][::-1],
             color="#D85A30", alpha=0.85)
axes[0].set_title("🚨 Mots → SPAM")
axes[0].set_xlabel("Score log P(spam) − log P(ham)")

axes[1].barh(feature_names[top_ham_idx][::-1], np.abs(log_probs[top_ham_idx][::-1]),
             color="#378ADD", alpha=0.85)
axes[1].set_title("✅ Mots → HAM")
axes[1].set_xlabel("Score |log P(spam) − log P(ham)|")

plt.tight_layout()
plt.savefig("output/06_top_features.png", dpi=150, bbox_inches="tight")
plt.show()
````

````
# ── Test sur de nouveaux messages ──────────────────────────
print("=" * 60)
print("  TEST EN TEMPS RÉEL — Nouveaux messages")
print("=" * 60)

test_messages = [
    "WINNER!! You've been selected as a lucky winner. Call now to claim your £1000 prize!",
    "Hey, are you coming to the study group tomorrow at 6pm?",
    "FREE entry in 2 a weekly competition to WIN an iPad 3G. Text WIN to 80085",
    "Can you please send me the homework for chapter 5?",
    "Congratulations! You have won a FREE iPhone. Click the link to claim it now!",
    "Ok I understand. See you later then.",
    "URGENT! Your mobile no 07808726822 was awarded a £2000 Bonus Caller Prize!",
]

print(f"  {'MESSAGE':<55} {'RÉSULTAT':<10} {'PROBA'}")
print("  " + "─" * 75)

for msg in test_messages:
    clean    = preprocess(msg)
    pred     = best_model.predict([clean])[0]
    prob     = best_model.predict_proba([clean])[0][1]
    label    = "🚨 SPAM" if pred == 1 else "✅ HAM "
    bar      = "█" * int(prob * 20) + "░" * (20 - int(prob * 20))
    print(f"  {msg[:55]:<55} {label}  {prob:.0%}")
````
